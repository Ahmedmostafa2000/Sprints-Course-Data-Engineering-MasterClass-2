```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data_filled_1.csv")
```

```python
df.head()
```

```python
gender_counts = df['Gender'].value_counts()

gender_counts.plot(kind='bar', color=['blue', 'pink'])

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender')

# Show the plot
plt.show()
```

```python
maritial_status_counts = df['Married'].value_counts()

maritial_status_counts.plot(kind='bar', color=['blue', 'gold'])

# Add labels and title
plt.xlabel('Married')
plt.ylabel('Count')
plt.title('Distribution of Maritial status')

# Show the plot
plt.show()
```

```python
Dependents_counts = df['Dependents'].value_counts()

Dependents_counts.plot(kind='bar',color='black')

# Add labels and title
plt.xlabel('Dependents')
plt.ylabel('Count')
plt.title('Distribution of Dependents')

# Show the plot
plt.show()
```
